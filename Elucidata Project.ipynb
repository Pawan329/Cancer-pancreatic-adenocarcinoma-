# Visualization using PCA
- We'll using genes data for dimension reduction and visualization

```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler

%matplotlib inline
```

### Transposing gene data:
- Now, Rows = Patient
-   columns = type of genes 

```python
Genes_data = Genes_data.transpose()
```

```python
Genes_data.shape
```

```python
Genes_data.head()
```

### Before PCA we always should do scaling:
- Is helps to normalise the data within a particular range

```python
scaler = StandardScaler()
```

```python
scaler.fit(Genes_data)
```

```python
scaled_data = scaler.transform(Genes_data)
```

```python
scaled_data[0:5]
```

### importing PCA package frok sklearn

```python
from sklearn.decomposition import PCA
```

### Keeping n_components = 3, it will keep 3 columns only with maximum information

```python
pca = PCA(n_components=3)
```

```python
pca.fit(scaled_data)
```

```python
x_pca = pca.transform(scaled_data)
```

```python
x_pca.shape
```

```python
type(x_pca)
```

```python
df_pca = pd.DataFrame(x_pca,columns=['PCA_1', 'PCA_2', 'PCA_3'])
```

```python
df_pca.head()
```

### Plotting Scatterplot to visualize data after applying PCA

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df_pca)
```

### Variance of Data after applying PCA

```python
df_pca.var()
```
